```python
df = pd.read_csv(os.path.join(img_path, "sample_labels.csv"))
df.head()
```

```python
df.rename(columns={"img-file": "file", "Finding Labels": "labels"}, inplace=True)
df.columns
```

```python
folders = np.unique(list(chain(*df["labels"].map(lambda x: x.split("|")).tolist())))
folders
```

```python
files_to_move = df.file.unique()
files_to_move
```

```python
# img_path = os.path.join(img_path, 'images')
img_path
```

### Create folders with `os.mkdir(path, mode)`

```python
for folder in folders:
    folder_path = os.path.join(img_path, folder)
    if not folder_path:
        os.mkdir(folder_path)

os.listdir(os.path.join(img_path))
```

```python
from tqdm import tqdm
```

```python
for row in tqdm(df.iterrows()):
    file = row[1]['file']
    label = row[1]['labels']
    try: 
        origin = os.path.join(img_path, file)
        
        if len(label.split('|')) > 1:
            folder = 'multilabel'

        else:
            folder = label
            
        destination = os.path.join(img_path, folder, file)
        shutil.move(origin, destination)
    except:
        pass
# destination = os.path.join(img_path, folder, file)
# shutil.move(origin, destination)
```
